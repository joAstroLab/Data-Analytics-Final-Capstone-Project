### **2. Data Cleaning & Pre-processing**

a.Dropping column:
Removing unnecessary column which is no longer required

```python
df = df.drop(columns=['torque'])
print(df.shape)
```

b.Removing duplicates: Removing duplicates improves the overall data quality, ensuring that insights and visualizations derived from the dataset are accurate.

```python
#Removing duplicates:
df = df.drop_duplicates()
df.duplicated().sum()
```

c.Convert data types: Changing columns into their proper formats such as text strings into numbers to ensure easiness.

```python
#Convert mileage to numeric column:
df['mileage'] = df['mileage'].astype(str).str.replace(' kmpl', '', regex=False)
df['mileage'] = pd.to_numeric(df['mileage'], errors='coerce')

#Convert engine to numeric column:
df['engine'] = df['engine'].astype(str)
df['engine'] = df['engine'].str.lower().str.replace(' cc', '', regex=False)
df['engine'] = pd.to_numeric(df['engine'], errors='coerce')

#Convert max_power to numeric column:
df['max_power'] = df['max_power'].astype(str).str.replace(' bhp', '', regex=False)
df['max_power'] = pd.to_numeric(df['max_power'], errors='coerce')
```

d.Handling missing values: Filling out mileage, engine and max power columns with median and seat column using mode for ensuring the data is complete.

```python
#Fill mileage with it's median:
df['mileage'] = df['mileage'].fillna(df['mileage'].median())

#Fill engine with it's median:
engine_median = df['engine'].median()
if pd.isna(engine_median):

  df['engine'] = df['engine'].fillna(0)
else:
  df['engine'] = df['engine'].fillna(engine_median)

#Fill max_power with it's median:
df['max_power'] = df['max_power'].fillna(df['max_power'].median())

#Fill seats with mode:
df['seats'] = df['seats'].fillna(df['seats'].mode()[0])
```

```python
#Verify if missing values are handled:
print(df.isnull().sum())
```

e.Extraction of brand name: To simplify the analysis and enable brand-wise camparisons, the brand name is extracted by splitting the text and selecting the first word. Later the original column is removed.

```python
#Extract first word:
df['brand'] = df['name'].str.split().str[0]
```

```python
#Drop the original name column:
df.drop(columns=['name'],inplace=True)
```

In these step, the dataset was cleaned and prepared for further analysis. Unnecessary columns were removed by filtering out the data, duplicate records were removed to ensure the data quality, data types of columns were converted for accurate analysis, missing values were handled using appropriate techniques such as median and mode. Additionally, irrelevant or inconsistent data entries were filtered out to improve the reliability of the results and compatibility.
